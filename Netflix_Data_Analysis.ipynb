```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Read the uploaded file
df = pd.read_csv('netflix_titles.csv')
```

```python
#checking for the null value in the datframe
df.isnull().sum()
```

```python
#forward fill the null values
df.fillna(method='ffill', inplace=True)
```

```python
df.isnull().sum()
```

```python
# Drop a column by name
df.drop('show_id', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop(['cast', 'description',''], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Save the DataFrame to an Excel file
df.to_excel(', index=False)
```

```python
# Distribution of content types
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='type')
plt.title('Distribution of Content Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(7, 6))
sns.countplot(data=df, x='release_year', order=sorted(df['release_year'].unique()))
plt.title('Number of Titles Released by Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert duration to numeric values (only for movies, not TV shows)
df['duration_numeric'] = df['duration'].apply(lambda x: int(x.split(' ')[0]) if 'min' in str(x) else None)

plt.figure(figsize=(7, 6))
sns.histplot(df[df['duration_numeric'].notnull()], x='duration_numeric', bins=30, kde=True)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
top_countries = df['country'].value_counts().nlargest(10).index
sns.countplot(data=df[df['country'].isin(top_countries)], y='country', order=top_countries)
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization 1: Distribution of Content Types
plt.figure(figsize=(5, 3))
sns.countplot(data=df, x='type')
plt.title('Distribution of Content Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

# Visualization 2: Distribution of Ratings
plt.figure(figsize=(5, 3))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Visualization 3: Number of Titles Released by Year
plt.figure(figsize=(5, 3))
sns.countplot(data=df, x='release_year', order=sorted(df['release_year'].unique()))
plt.title('Number of Titles Released by Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Visualization 4: Duration of Movies
df['duration_numeric'] = df['duration'].apply(lambda x: int(x.split(' ')[0]) if 'min' in str(x) else None)
plt.figure(figsize=(5, 3))
sns.histplot(df[df['duration_numeric'].notnull()], x='duration_numeric', bins=30, kde=True)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()

# Visualization 5: Top 10 Countries by Number of Titles
plt.figure(figsize=(5, 3))
top_countries = df['country'].value_counts().nlargest(10).index
sns.countplot(data=df[df['country'].isin(top_countries)], y='country', order=top_countries)
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```
